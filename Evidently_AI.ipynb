We can even take the dataset and exactly mention what test to be performed on particular columns.

```python
import pandas as pd
from evidently import ColumnMapping
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset
from scipy.stats import ks_2samp

# Load and preprocess your datasets

# Use evidently for data drift analysis
current_data=iris_frame.iloc[:60]
reference_data=iris_frame.iloc[60:]
data_drift_report = Report(metrics=[DataDriftPreset()])
data_drift_report.run(current_data=current_data, reference_data=reference_data, column_mapping=None)

# Extract relevant data for KS test (e.g., two samples)
sample1 = current_data['sepal width (cm)'].values
sample2 = reference_data['sepal width (cm)'].values

# Perform KS test
ks_statistic, p_value = ks_2samp(sample1, sample2)

# Print results
print("KS Statistic:", ks_statistic)
print("P-value:", p_value)

# Display evidently report
data_drift_report.show()
```

# **TargetDriftPreset**

```python
import pandas as pd
from sklearn.datasets import load_iris
from evidently import ColumnMapping
from evidently.report import Report
from evidently.metric_preset import TargetDriftPreset

# Load the Iris dataset
iris = load_iris()
iris_frame = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_frame['target'] = iris.target

# Split the dataset into current and reference data
current_data = iris_frame.iloc[:60]
reference_data = iris_frame.iloc[60:]

# Define column mapping (optional, but useful for specifying the target column)
column_mapping = ColumnMapping(target='target')

# Create the target drift report
target_drift_report = Report(metrics=[
    TargetDriftPreset(),
])

# Run the report
target_drift_report.run(current_data=current_data, reference_data=reference_data, column_mapping=column_mapping)

# To display the report in Jupyter Notebook
target_drift_report
```

# **DataDriftTestPreset**

Setting monitoring and alerts through emails using smtp when datadrift is detected.

```python
import pandas as pd
from sklearn.datasets import load_wine
from evidently.test_suite import TestSuite
from evidently.test_preset import DataDriftTestPreset
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Load Wine dataset
wine_data = load_wine(as_frame=True)
wine_frame = wine_data.frame

# Simulate current data by shuffling the original data and changing the target distribution slightly
current_data = wine_frame.sample(frac=1.0, random_state=42).reset_index(drop=True)
current_data['target'] = current_data['target'].apply(lambda x: (x + 1) % 3)
# Modify the target to simulate drift

# Initialize the data drift test suite focusing on the target column
data_drift_tests = TestSuite(tests=[
    DataDriftTestPreset()
])

# Run the data drift tests, specifying the target column
data_drift_tests.run(reference_data=wine_frame[['target']], current_data=current_data[['target']])

# Generate and print the report
drift_report = data_drift_tests.as_dict()
print(drift_report)

# Extract the share of drifted features
total_columns_count = len(wine_frame.columns)
failed_tests = [test for test in drift_report['tests'] if test['status'] == 'FAIL']
share_drifted_features = len(failed_tests) / total_columns_count

# Check if the drift exceeds the threshold and set an alert
drift_threshold = 0.1  # 10%
if share_drifted_features > drift_threshold:
    print("Alert: Significant data drift detected!")
    # Email details
    sender_email = "kushmithachoudary123@gmail.com"
    receiver_email = "kusumitha.c@sigmoidanalytics.com"
    password = "*************"  # or your app-specific password
    # Create the email content
    message = MIMEMultipart("alternative")
    message["Subject"] = "Data Drift Alert"
    message["From"] = sender_email
    message["To"] = receiver_email

    # Email body
    text = f"Significant data drift detected! The share of drifted features is {share_drifted_features:.2%}."
    part = MIMEText(text, "plain")
    message.attach(part)

    # Send the email
    try:
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(sender_email, password)
            server.sendmail(sender_email, receiver_email, message.as_string())
        print("Alert email sent!")
    except Exception as e:
        print(f"Failed to send email: {e}")
else:
    print("No significant data drift detected.")
    print("No alert email sent.")
```
